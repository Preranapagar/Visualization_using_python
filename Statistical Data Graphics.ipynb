# <span style='color:orange '> Creating Statistical data graphics </span>

```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize']=5,4
sns.set_style('whitegrid')    
```

### Histograms:

* A histogram shows a variable's distribution as set of adjacent rectangles on a data chart. histogram represnts counts of data withing
a numeric range of values

```python
cars=pd.read_csv('autos_dataset.csv')
cars.head(5)
```

```python
cars.columns
```

```python
cars.index=cars.make
```

```python
mpg=cars['city-mpg']
mpg.plot(kind='hist') #direct plotting
```

```python
higway_mpg=cars['highway-mpg']
plt.hist(higway_mpg) #ploting by calling function
plt.plot()
```

```python
#histogram using seaborn library

sns.distplot(mpg)
```

### Scatterplots:

* Scatterplots are useful when you want to explore interrelations and dependencies between two
different variables. These data graaphics are ideal for visually spotting outliers and trends in data

```python
cars_df=cars[:30]
cars_df.plot(kind='scatter',x='bore',y='stroke',c='grey',s=150)
```

```python
#using seaborn
sns.regplot(x='width',y='height',scatter=True,data=cars_df)
```

### Scatterplot Matrices:

* Scatterplot matrices use to visually display correlation between two or more variables

```python
from sklearn import datasets
iris=datasets.load_iris()
iris=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

### Boxplots:

* Useful for seeing variables spread and detecting outliers

```python
cars.boxplot(column='city-mpg',by='fuel-type')
```

```python
sns.boxplot(x='fuel-type',y='city-mpg',data=cars,palette='hls') #using seaborn
```
