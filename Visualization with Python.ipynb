## Data Visualization using Python

###### Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris,load_wine
```

###### Datasets

```python
#we are using iris and wine dataset
```

```python
iris=load_iris()
iris_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
iris_df['target']=iris.target
```

```python
wine=load_wine()
wine_df=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_df['target']=wine.target
```

```python
iris_df.columns
```

```python
iris_df.rename({'sepal length (cm)':'sepal_length', 'sepal width (cm)':'sepal_width', 'petal length (cm)':'petal_length',
       'petal width (cm)':'petal_width'},inplace=True,axis=1)
iris_df.columns
```

### 1. Scatter Plot

```python
#create figure and axis
fig,ax=plt.subplots()

#scatter the sepal_length against sepal_width

ax.scatter(iris_df['sepal_length'],iris_df['sepal_width'])

#set titles and lables

ax.set_title('Iris Dataset')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
```

```python
#create colour dictionary

colors={0:'r',1:'g',2:'b'}

#create figure and axis
fig, ax=plt.subplots()

#plot each datapoint
for i in range(len(iris_df['sepal_length'])):
    ax.scatter(iris_df['sepal_length'][i],iris_df['sepal_width'][i],color=colors[iris_df['target'][i]])
    
#set titles and lables
ax.set_title('Iris Dataset')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal width')
```

```python
iris_df.plot.scatter(x='sepal_length',y='sepal_width',title='Iris Dataset')
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='target',data=iris_df)
```

## 2. Line Chart

```python
#get columns to plot
columns=iris_df.drop(['target'],axis=1)

#create x data
x_data=range(0,iris_df.shape[0])

#create fig and axis
fig,ax=plt.subplots()

#plot each column
for column in columns:
    ax.plot(x_data,iris_df[column],label=column)
    
#set titles and legends
ax.set_title('Iris Dataset')
ax.legend()
```

```python
iris_df.drop('target',axis=1).plot.line(title='Iris dataset')
```

```python
sns.lineplot(data=iris_df.drop('target',axis=1))
```

## 4. Bar Chart

```python
#create figure and axis
fig, ax=plt.subplots()

#count the occurance
data=wine_df['magnesium'].value_counts()

#get x and y data
points=data.index
frequency=data.values

#creat bar chart
ax.bar(points,frequency)

#sel title and lables
ax.set_title('Magnesium Content')
```

## 6. Heatmap

```python
#get correlation matrix
corr=iris_df.corr()

fig, ax=plt.subplots()
im=ax.imshow(corr.values)

#set lables
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

#rotate ticks label
plt.setp(ax.get_xticklabels(),rotation=45,ha='right', rotation_mode='anchor')
```

```python
#get correlation matrix
corr=iris_df.corr()

fig, ax=plt.subplots()
im=ax.imshow(corr.values)

#set labels

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

#set rotation

plt.setp(ax.get_xticklabels(),rotation=45)

#loop over data dimensions to create text annotations

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text=ax.text(j,i,np.around(corr.iloc[i,j],decimals=2), ha='center',va='center',color='black')
```

```python
#create with seaborn with easier way

sns.heatmap(iris_df.corr(),annot=True)
```
