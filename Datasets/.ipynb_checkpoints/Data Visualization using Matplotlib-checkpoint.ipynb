```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('autos_dataset.csv')
df.head()
```

## 1.Histogram

A histogram takes in a series of data and divides the data into a number of bins. It then plots the frequency data points in each bin (i.e. the interval of points). It is useful in understanding the count of data ranges.

When to use: We should use histogram when we need the count of the variable in a plot.

```python
plt.hist(df['city-mpg'],
        facecolor='peru',
        bins=10)
plt.show()
```

```python

plt.hist(df['city-mpg'],facecolor='grey',bins=10)
plt.show()
```

components of histogram:
    n: contains a frequency of each bin
    bins: Represents the middle value of each bin
    patches: the patch object for the rectangle shape representing each bar

```python
n, bins, patches =plt.hist(df['city-mpg'],
                          facecolor='peru',
                          edgecolor='black',
                          bins=10)
print('n:',n)
print('bins:',bins)
print('patches:',patches)
```

```python
#to plot cumulative plot
plt.hist(df['city-mpg'],
         facecolor='peru',
         edgecolor='black',
         bins=10,
        cumulative=True)
plt.show()
```

##### Multiple Histogram

```python
plt.hist(df['city-mpg'],
        facecolor='orangered',
        edgecolor='maroon',
        bins=8)

plt.hist(df['highway-mpg'],
        facecolor='peru',
        edgecolor='blue',
        bins=10,
        alpha=0.3)
plt.show()
```

## 2. Pie Chart

```python
plt.pie(df['fuel-type'].value_counts(), labels=df['fuel-type'].unique())
plt.axis('equal')
plt.show()
```

components of pie chart:
    wedges:a list of patch objects representing each wedge
    texts:list of text objects representing the labels
    autotexts:list of text object for the numeric values

```python
wedges, text, autotexts =plt.pie(df['fuel-type'].value_counts(),labels=df['fuel-type'].unique(),autopct='%.2f')
plt.axis('equal')

print('wedges:',wedges)
print('texts:',text)
print('autotexts:',autotexts)
```

```python
#chart customization

colors=['darkorange','sandybrown']
plt.pie(df['fuel-type'].value_counts(), 
        labels=df['fuel-type'].unique(),
        colors=colors,
        autopct='%.2f')

plt.axis('equal')
plt.show()
```

```python
#to explode chart

explode=(0,0.1,0,0.2,0.3)
colorsp=['darkorange','sandybrown','darksalmon','orangered','chocolate']
plt.pie(df['body-style'].value_counts(),
       labels=df['body-style'].unique(),
       colors=colorsp,
       explode=explode,
       autopct='%.2f')

plt.show()
```

## 3. Box Plot

```python
plt.boxplot(df['wheel-base'],
           patch_artist=True,
           notch=True)
plt.show()
```

```python
exams=pd.read_csv('exams.csv')
exams.head()
```

```python
score_card=exams[['math score','reading score','writing score']]
score_card.head()
```

```python
score_card.describe()
```

```python
#for ploting boxplot we convert data to array
score_array=np.array(score_card)
score_array
```

```python
colors=['blue','green','grey']

bp=plt.boxplot(score_array,
              patch_artist=True,
              notch=True)

for i in range(len(bp['boxes'])):
    bp['boxes'][i].set(facecolor=colors[i])
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1,2,3],['Math','Reading','Writing'])
plt.show()
```

## 6. TwinAxis

TwinAxis helps in visualizing plotting 2 plots w.r.t to the y-axis and same x-axis.
When to use: It should when we require 2 plots or grouped data in the same direction.

Eg: Population, GDP data in the same x-axis (Date).

```python
austin_weather=pd.read_csv('Datasets/austin_weather.csv')
austin_weather.head()
```

```python
austin_weather=austin_weather[['Date','TempAvgF','WindAvgMPH']].head(30)
austin_weather
```

```python
#subplot for ploting figures
fig,ax_tempf=plt.subplots()

#similar to fig=plt.figure(figsize(12,6))
fig.set_figwidth(12)
fig.set_figheight(6)

#set x label which is common
ax_tempf.set_xlabel('Date')

#bottom=false disable ticks and label bottom disable x-axis labels
ax_tempf.tick_params(axis='x',
                    bottom=False,
                    labelbottom=False)

#set left y axis label
ax_tempf.set_ylabel('Temp(F)',
                   color='red',
                   size='large')

#set label color and label size for the left y-axis
ax_tempf.tick_params(axis='y',
                    labelcolor='red',
                    labelsize='large')

#plot avg temp on y-axis on left
ax_tempf.plot(austin_weather['Date'],
             austin_weather['TempAvgF'],
             color='red')

#twinx sets the same x-axis for both the plots
ax_precip=ax_tempf.twinx()

#set right y-axis label
ax_precip.set_ylabel('Avg Wind speed',
                    color='blue',
                    size='x-large')

#set label color and label size to right y-axis
ax_precip.tick_params(axis='y',
                     labelcolor='blue',
                     labelsize='large')

#plot windAvg on y-axis to the right
ax_precip.plot(austin_weather['Date'],
              austin_weather['WindAvgMPH'],
              color='blue')

#ax_precip.legend(0)

plt.show()
```

### Twin axis for the same data

if same data can be expressed in different units, we can configure our axis to display both units on two axes

```python
def fahrenheit2celcius(f):
    return (f-32)*5/9
```

```python
#subplot for plotting figure
fig, ax=plt.subplots()
fig=plt.figure(figsize=(12,6))

#set x label which is common
ax.set_xlabel('Date')

#disable bottom ticks and labels
ax.tick_params(axis='x',
              bottom=False,
              labelbottom=False)

#set left y-axis label
ax.set_ylabel('Temp(F)',color='red',size='x-large')

#set label color and label size on left y-axis
ax.tick_params(axis='y',
              labelcolor='red',
              labelsize='large')

#plot AvgTemp on y-axis to the left
ax.plot(austin_weather['Date'],
       austin_weather['TempAvgF'],
       color='green')

#twinx set the same axis for both plot
ax_t=ax.twinx()

#get ymin and ymax of AvgTemp in Fahrenheit
ymin, ymax = ax.get_ylim()

#set ymin and ymin of AvgTemp in Fahrenheit to AvgTemp in celsius
ax_t.set_ylim((fahrenheit2celcius(ymin),fahrenheit2celcius(ymax)))


#set right y-axis label
ax_t.set_ylabel('Temp(C)',
               color='blue',
               size='x-large')

#set label color and label size on right y-axis
ax_t.tick_params(axis='y',
                labelcolor='blue',
                labelsize='large')
plt.show()
```

## 7. Stack Plot and Stem Plot

Stack Plot : visualizes dat ain stacks and shows distribution of data over time
    
when to use: It is used for checking multiple variable area plots in a single plot
    
eg. It is useful in understanding the change of distribution in multiple variables over an interval

```python
np_data=pd.read_csv('Datasets/national_parks.csv')
np_data.head()
```

```python
x=np_data['Year']
y=np.vstack([np_data['Badlands'],
            np_data['GrandCanyon'],
            np_data['BryceCanyon']])
```

```python
#label for each stack
labels=['Badlands','GrandCanyon','BryceCanyon']

#color for each stack
colors=['sandybrown','tomato','skyblue']

#similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

#plot legends to upperleft of figure
plt.legend(loc=2)
plt.show()
```

#### Precentage stack area chart

stack plots based on percentage in distribution

```python
plt.figure(figsize=(10,7))

data_prep=np_data.divide(np_data.sum(axis=1),axis=0)

plt.stackplot(x,data_prep['Badlands'],data_prep['GrandCanyon'],data_prep['BryceCanyon'],
             edgecolor='black',
             colors=colors,
             labels=labels)

plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.show()
```

Stem plot: similar to stack plot but helps to view negative numbers, i.e difference in data over time

```python
np_stem=np_data.copy()
```

```python
np_stem[['Badlands','GrandCanyon','BryceCanyon']]=np_data[['Badlands','GrandCanyon','BryceCanyon']].diff()
np_stem.head()
```

```python
plt.figure(figsize=(15,11))

plt.suptitle('Change in number of visitors')

plt.subplot(311)
plt.stem(np_stem['Year'],
        np_stem['Badlands'],
        markerfmt='b_',
        linefmt='r--',
        basefmt='g:')
plt.title('Badlands')

plt.subplot(312)
plt.stem(np_stem['Year'],
        np_stem['GrandCanyon'],
        markerfmt='b_',
        linefmt='r--',
        basefmt='g:')
plt.title('GrandCanyon')

plt.subplot(313)
plt.stem(np_stem['Year'],
        np_stem['BryceCanyon'],
        markerfmt='b_',
        linefmt='r--',
        basefmt='g:')
plt.title('BryceCanyon')

plt.show()
```

## 8. Bar Plot

Bar Plot shows the distribution of data over several groups. It is commonly confused with a histogram which only takes numerical data for plotting. It helps in comparing multiple numeric values.

When to use: It is used when to compare between several groups.

Eg: Student marks in an exam.

```python
countries=pd.read_csv('Datasets/countries.csv')
countries.head()
```

```python
data_2007=countries[countries['year']==2007]
top_10=data_2007.sort_values('population',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))

x=range(10)
plt.bar(x,top_10['population']/10**9)
plt.xticks(x,top_10['country'])
plt.xlabel('countries')
plt.ylabel('population in billions')
plt.title('10 most populated countries in 2007')
plt.show()
```

## 9.Scatter plot

Scatter plot helps in visualizing 2 numeric variables. It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. It also helps in detecting outliers in the plot.

When to use: It is used in Machine learning concepts like regression, where x and y are continuous variables. It is also used in clustering scatters or outlier detection.

```python
plt.scatter(data_2007['gdpPerCapita'],
           data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')

plt.show()
```

```python
data_2007['gdpPerCapita'].corr(data_2007['lifeExpectancy'])
```

```python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
np.log10(data_2007['gdpPerCapita']).corr(data_2007['lifeExpectancy'])
```
