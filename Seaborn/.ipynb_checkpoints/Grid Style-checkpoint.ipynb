```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.feature_names
```

```python
df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
```

```python
df['Target']= boston.target
```

```python
df
```

```python
sns.set_style('darkgrid')
```

```python
sns.displot(data=df,kind='kde',x='AGE',y='Target')
```

```python
sns.set_style('whitegrid')
```

```python
sns.displot(data=df,kind='kde',x='AGE',y='Target')
```

```python
sns.set_style('dark')
sns.displot(data=df,kind='kde',x='AGE',y='Target')
```

```python
sns.set_style('white')
sns.displot(data=df,kind='kde',x='AGE',y='Target')
```

```python
sns.set_style('ticks')
sns.displot(data=df,kind='kde',x='AGE',y='Target')
```
